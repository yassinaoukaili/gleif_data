### Search LEIs by filters

```python
filters = LeiRecordFilter(legal_form='CDOV', country='AG', status='LAPSED')
GleifData.get_companies_df(filters)
```

```python
# Alternatively obtain company objects for OOP purposes 
companies = GleifData.get_companies(filters)

for company in companies:
    print(company.legalName.name, company.lei)
```
